## Escenario:
Tu tarea es escribir un simple programa que simule jugar a tic-tac-toe (nombre en inglés) con el usuario. Para hacerlo más fácil, hemos decidido simplificar el juego. Aquí están nuestras reglas:

* La maquina (por ejemplo, el programa) jugará utilizando las 'X's.
  
* El usuario (por ejemplo, tu) jugarás utilizando las 'O's.
  
* El primer movimiento es de la maquina: siempre coloca una 'X' en el centro del tablero.

* Todos los cuadros están numerados comenzando con el 1 (observa el ejemplo para que tengas una referencia).

* El usuario ingresa su movimiento introduciendo el número de cuadro elegido. El número debe de ser valido, por ejemplo un valor entero mayor que 0 y menor que 10, y no puede ser un cuadro que ya esté ocupado.

* El programa verifica si el juego ha terminado. Existen cuatro posibles veredictos: el juego continua, el juego termina en empate, tu ganas, o la maquina gana.

* La maquina responde con su movimiento y se verifica el estado del juego.

* No se debe implementar algún tipo de inteligencia artificial, la maquina elegirá un cuadro de manera aleatoria, eso es suficiente para este juego.

## Requerimientos:
Implementa las siguientes características:

* El tablero debe ser almacenado como una lista de tres elementos, mientras que cada elemento es otra lista de tres elementos (la lista interna representa las filas) de manera que todos los cuadros puedas ser accedidos empleado la siguiente sintaxis: board[row][column]

* Cada uno de los elementos internos de la lista puede contener 'O', 'X', o un digito representando el número del cuadro (dicho cuadro se considera como libre).
  
* La apariencia de tablero debe de ser igual a la presentada en el ejemplo.

* Implementa las funciones definidas para ti en el editor.

Para obtener un valor numérico aleatorio se puede emplear una función integrada de Python denominada randrange(). 

## Solución:

```python
from random import randint

ganador = False

tablero = [[1, 2, 3],
           [4, 5, 6],
           [7, 8, 9]]

lista_X = []
lista_O = []

jugador = 1 #1 es la maquina, 2 el usuario

def vizualizar_tablero():

    #print( "\n--- Juego del gato ---")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print(f"|   {tablero[0][0]}   |   {tablero[0][1]}   |   {tablero[0][2]}   |")
    print("|       |       |       |")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print(f"|   {tablero[1][0]}   |   {tablero[1][1]}   |   {tablero[1][2]}   |")
    print("|       |       |       |")
    print("+-------+-------+-------+")
    print("|       |       |       |")
    print(f"|   {tablero[2][0]}   |   {tablero[2][1]}   |   {tablero[2][2]}   |")
    print("|       |       |       |")
    print("+-------+-------+-------+\n")


def tiro():

    global jugador

    entro = False

    while entro == False:

        try:

            if jugador == 1:
                print("\nTurno maquina:")
                numero = randint(1, 10)
                valor = "O"
            
            if jugador == 2:
                print("\nTurno usuario:")
                
                while True:
                    numero = int(input("Ingresa el numero de la casilla para colocar 'X': "))

                    if numero > 9 or numero < 1:
                        print("Error ingresa un valor entre 1 y 9 porfavor..")
                    else:
                        break


                valor = "X"

            posicion = 0

            for fila in range(len(tablero)):

                for columna in range(len(tablero)):
                    posicion += 1
                    
                    if numero == posicion:
                        if tablero[fila][columna] == "X" or tablero[fila][columna] == "O":
                            if jugador == 2:
                                print("Error ese espacio ya esta ocupado")
                                continue
                            else:
                                continue
                        else:

                            #cambiando el valor de la lista tablero en la posicion de tablero[fila][columna] por X - O
                            tablero[fila][columna] = valor
                            entro = True 

                            #Agregando las posiciones de "X" - "O" a sus listas
                            if valor == "X":
                                lista_X.append(posicion)
                            
                            elif valor == "O":
                                lista_O.append(posicion)
                            else:
                                pass
                            
                            vizualizar_tablero()
                            comprobando_ganador()

                            #Cambiando de jugador al tirar 
                            if jugador == 1:
                                jugador = 2
                            else:
                                jugador = 1
                            break
                    else:
                        pass
        
        except ValueError:
            print("Error ingrese el numero de la casilla en el formato correcto")




def comprobando_ganador():

    global ganador
    global jugador
    
    #buscando si hay ganador
    if jugador == 1:
        name = "maquina"
        lista = lista_O
    
    if jugador == 2:
        name = "usuario"
        lista = lista_X
    
    #filas
    if 1 in lista and 2 in lista and 3 in lista:
        print(f"Ganador jugador {name}")
        ganador = True 
    elif 4 in lista and 5 in lista and 6 in lista:
        print(f"Ganador jugador {name}")
        ganador = True
    elif 7 in lista and 8 in lista and 9 in lista:
        print(f"Ganador jugador {name}")
        ganador = True
    
    #columnas
    if 1 in lista and 4 in lista and 7 in lista:
        print(f"Ganador jugador {name}")
        ganador = True 
    elif 2 in lista and 5 in lista and 8 in lista:
        print(f"Ganador jugador {name}")
        ganador = True
    elif 3 in lista and 6 in lista and 9 in lista:
        print(f"Ganador jugador {name}")
        ganador = True
    
    #diagonales
    if 1 in lista and 5 in lista and 9 in lista:
        print(f"Ganador jugador {name}")
        ganador = True 
    elif 3 in lista and 5 in lista and 7 in lista:
        print(f"Ganador jugador {name}")
        ganador = True
        

    #validando espacios libres para continuar o marcar empate
    if ganador == False:

        espacios = False

        for fila in range(len(tablero)):

            for columna in range(len(tablero)):
            
                if tablero[fila][columna] != "X" and tablero[fila][columna] != "O":
                    espacios = True
        
        if espacios == False:
            print("Empate saliendo del programa")
            exit()   


    return ganador



def main():

    global ganador

    while ganador == False:
        tiro()
        
    print("Saliendo del programa")
    exit()


main()
```
